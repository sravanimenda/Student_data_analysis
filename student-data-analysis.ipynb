```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\srava\\OneDrive\\Desktop\\Jupyter-projects\\studentmarks1.csv")
```

```python
df["total_marks"]=df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
```

```python
print(df)
```

```python
df["percentage"]=((df["total_marks"]/600)*100).round(2)
```

```python
print(df)
```

```python
def assign_grade(mark):
    if mark >= 75:
        return 'Distinction'
    elif 65 <= mark < 75:
        return 'First Class'
    elif 60 <= mark < 65:
        return 'Second Class'
    else:
        return 'Third Class'

# Apply function to calculate grade
df['grade'] = df['percentage'].apply(assign_grade)
```
